The following code visualizes the data in a series of bar charts in order to find trends and biases. After running, the bar charts will display on the notebook itself. I chose to display it rather than save it to the local computer in order to save space and make the data set easily updatable without requiring even more images to be saved to the computer. I compared each of the variables with the final result we will be trying to predict: the order.

As we can see, orders are almost entirely comprised of sophomores and juniors. Assumedly, this would be because freshmen have a meal plan to use, and seniors are cooking for themselves in an off campus apartment.

There is relatively high variability based on major, but the majors towards the left side of the graph tend to order more food from these trucks overall. This is also the case for universities, however there are notable outliers. Specifically, Purdue, IU Bloomington, and DePauw tend to order very little to no food from these trucks. For both of these cases, there is a large chance of bias: food trucks may not be on certain campuses as much, or in certain areas of campus where certain majors may frequent. 

The time where most orders are placed is around lunchtime, specifically in the 11:00 am to 2:00 pm range. This makes sense, as most people eat at designated meal times in between their classes as close to a regular lunch time as possible. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dataset.csv').dropna()

frequencies = []
orders = data.Order.unique().tolist()
times = data.Time.unique().tolist()
times.sort()
universities = data.University.unique().tolist()
majors = data.Major.unique().tolist()
years = data.Year.unique().tolist()
years.sort()

#create graphs to analyze distribution
for i in orders:
    
    for j in years:
        mask = (data.Order == i) & (data.Year == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(years, frequencies)
    plt.xlabel('Year')
    plt.ylabel('Frequency')
    plt.title(i + " ordered by each year")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()
    
    for j in majors:
        mask = (data.Order == i) & (data.Major == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(majors, frequencies)
    plt.xlabel('Major')
    plt.ylabel('Frequency')
    plt.title(i + " ordered by each major")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()
    
    for j in universities:
        mask = (data.Order == i) & (data.University == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(universities, frequencies)
    plt.xlabel('University')
    plt.ylabel('Frequency')
    plt.title(i + " ordered by each university")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()

    for j in times:
        mask = (data.Order == i) & (data.Time == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(times, frequencies)
    plt.xlabel('Time')
    plt.ylabel('Frequency')
    plt.title(i + " ordered at each time")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()
```
