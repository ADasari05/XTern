```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
from sklearn.tree import export_graphviz
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from scipy.stats import randint
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dataset.csv').dropna()

frequencies = []
orders = data.Order.unique().tolist()
times = data.Time.unique().tolist()
times.sort()
universities = data.University.unique().tolist()
majors = data.Major.unique().tolist()
years = data.Year.unique().tolist()
years.sort()

#create graphs to analyze distribution
for i in orders:
    
    for j in years:
        mask = (data.Order == i) & (data.Year == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(years, frequencies)
    plt.xlabel('Year')
    plt.ylabel('Frequency')
    plt.title(i + " ordered by each year")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()
    
    for j in majors:
        mask = (data.Order == i) & (data.Major == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(majors, frequencies)
    plt.xlabel('Major')
    plt.ylabel('Frequency')
    plt.title(i + " ordered by each major")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()
    
    for j in universities:
        mask = (data.Order == i) & (data.University == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(universities, frequencies)
    plt.xlabel('University')
    plt.ylabel('Frequency')
    plt.title(i + " ordered by each university")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()

    for j in times:
        mask = (data.Order == i) & (data.Time == j)
        count = len(data[mask])
        frequencies.append(count)
    plt.bar(times, frequencies)
    plt.xlabel('Time')
    plt.ylabel('Frequency')
    plt.title(i + " ordered at each time")
    frequencies = []
    plt.xticks(rotation = 90)
    plt.show()
```
